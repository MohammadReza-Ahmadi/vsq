```
import pandas as pd

filepath = '/content/drive/MyDrive/CreditScoring/Mellat data (10000).xlsx'
data = pd.read_excel(filepath)
print(data.shape)
```

```
# imports
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```
sns.set()
fig = data['sector'].value_counts().plot(kind='bar')
```

```
fig.figure.savefig('/content/figure.png')
fig.figure.savefig('/content/figure.svg')
```

```
sns.countplot(x="credit type", data=data, palette="Reds")
plt.show()
```

```
data['education'].value_counts().plot(kind='bar')
```

```
sns.countplot(x="gender", data=data, palette="Reds")
plt.show()
```

```
sns.displot(data, x="age ", kde=True)
```

```
sns.jointplot(x='interest rate',
              y='collateral amount ', data=data)
plt.show()
```

```
sns.jointplot(x='interest rate',
              y='collateral amount ',
              data=data, kind='hex')
plt.show()
```

```
sns.jointplot(x='age ',
              y='sector', data=data, kind='kde')
plt.show()
```

```
sns.set_style("ticks")
ax = sns.pairplot(data.drop(columns=['id']),
             hue='label', diag_kind="kde", kind="scatter", palette="husl")
plt.show()
```

```
ax.savefig('/content/figure.svg')
```

```
ax.savefig('/content/figure.png')
```

```
sns.stripplot(x="label", y="collateral type", data=data, jitter=True)
plt.show()
```

```
sns.boxplot(x="label",
            y="interest rate", data=data)
plt.show()
```

```
sns.violinplot(x="label", y="interest rate", data=data)
plt.show()
```

```
sns.violinplot(x="label", y="interest rate",
               hue='gender', data=data, dodge=True, split=True)
plt.show()
```

```
sns.violinplot(x="label", y="interest rate",
               hue='gender', data=data, dodge=True, split=False)
plt.show()
```

```
sns.boxplot(data=data.drop(columns=['id']), orient="h")
plt.show()
```

```
sns.violinplot(data=data.drop(columns=['id']), orient="w")
plt.show()
```

```
sns.factorplot(x="gender", y="interest rate", hue="label",
               kind='violin',data=data)
plt.show()
```

```
sns.factorplot("label", col="sector", col_wrap=3, data=data, kind="count")
```

```
sns.regplot(x="collateral amount ", y="credit amount", data=data, order=1)
sns.lmplot(x="collateral amount ", y="credit amount", data=data, order=1)
```

```
g = sns.FacetGrid(data, col="collateral type")
g.map(plt.hist, "label")
plt.show()
```

```
g = sns.FacetGrid(data, col="collateral type", hue="credit amount")
g.map(plt.scatter, "label", "age ")
plt.show()
```

```
g = sns.PairGrid(data.drop(columns=['id']))
g.map(plt.scatter)
plt.show()
```

```
g = sns.PairGrid(data.drop(columns=['id']))
g.map_diag(plt.hist)
g.map_offdiag(plt.scatter)
plt.show()
```

```
g = sns.PairGrid(data.drop(columns=['id']))
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot, cmap="Blues_d")
g.map_diag(sns.kdeplot, lw=3, legend=False)
plt.show()
```
