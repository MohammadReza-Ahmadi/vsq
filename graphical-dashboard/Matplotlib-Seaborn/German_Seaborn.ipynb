```
# load german credit data
import scorecardpy as sc
data = sc.germancredit()
```

```
# imports
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```
data.head(5)
```

```
data.describe()
```

```
data.info()
```

```
data.columns
```

```
sns.set()
data['purpose'].value_counts().plot(kind='bar')
```

```
sns.countplot(x="purpose", data=data, palette="Reds")
plt.show()
```

```
data['job'].value_counts().plot(kind='bar')
```

```
sns.countplot(x="job", data=data, palette="Reds")
plt.show()
```

```
sns.displot(data, x="credit.amount", kde=True)
```

```
sns.jointplot(x='age.in.years',
              y='number.of.existing.credits.at.this.bank', data=data)
plt.show()
```

```
sns.jointplot(x='age.in.years',
              y='number.of.existing.credits.at.this.bank', data=data, kind='hex')
plt.show()
```

```
sns.jointplot(x='age.in.years',
              y='number.of.existing.credits.at.this.bank', data=data, kind='kde')
plt.show()
```

```
sns.set_style("ticks")
sns.pairplot(data[['age.in.years', 'number.of.existing.credits.at.this.bank', 'creditability']],
             hue='creditability', diag_kind="kde", kind="scatter", palette="husl")
plt.show()
```

```
sns.stripplot(x="creditability", y="credit.amount", data=data, jitter=False)
plt.show()
```

```
sns.swarmplot(x="creditability", y="foreign.worker", data=data)
plt.show()
```

```
sns.boxplot(x="creditability",
            y="credit.amount", data=data)
plt.show()
```

```
sns.violinplot(x="creditability", y="credit.amount", data=data)
plt.show()
```

```
sns.violinplot(x="creditability", y="credit.amount",
               hue='foreign.worker', data=data, dodge=True, split=True)
plt.show()
```

```
sns.violinplot(x="creditability", y="credit.amount",
               hue='foreign.worker', data=data, dodge=True, split=False)
plt.show()
```

```
sns.barplot(x="credit.amount", y="creditability", hue="foreign.worker", data=data)
plt.show()
```

```
sns.countplot(x="creditability", data=data, palette="Reds")
plt.show()
```

```
sns.pointplot(x="credit.amount", y="foreign.worker", hue="creditability", data=data)
plt.show()
```

```
sns.boxplot(data=data, orient="h")
plt.show()
```

```
sns.violinplot(data=data, orient="w")
plt.show()
```

```
sns.factorplot(x="credit.amount", y="foreign.worker", hue="creditability", data=data)
plt.show()
```

```
sns.factorplot(x="credit.amount", y="foreign.worker", hue="creditability",
               kind='violin',data=data)
plt.show()
```

```
sns.factorplot("creditability", col="age.in.years", col_wrap=3, data=data, kind="count")
```

```
sns.regplot(x="age.in.years", y="credit.amount", data=data, order=5)
sns.lmplot(x="age.in.years", y="credit.amount", data=data, order=5)
```

```
g = sns.FacetGrid(data, col="foreign.worker")
g.map(plt.hist, "creditability")
plt.show()
```

```
g = sns.FacetGrid(data, col="foreign.worker", hue="credit.amount")
g.map(plt.scatter, "creditability", "age.in.years")
plt.show()
```

```
g = sns.PairGrid(data[['foreign.worker', 'age.in.years', 'credit.amount', 'creditability', ]])
g.map(plt.scatter)
plt.show()
```

```
g = sns.PairGrid(data[['foreign.worker', 'age.in.years', 'credit.amount', 'creditability', ]])
g.map_diag(plt.hist)
g.map_offdiag(plt.scatter)
plt.show()
```

```
g = sns.PairGrid(data[['foreign.worker', 'age.in.years', 'credit.amount', 'creditability', ]])
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot, cmap="Blues_d")
g.map_diag(sns.kdeplot, lw=3, legend=False)
plt.show()
```
